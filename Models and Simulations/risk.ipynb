```python
import pandas as pd

# Load CSV file
file_path = '/Users/mubeen/Library/CloudStorage/OneDrive-TheUniversityofHongKong/Year5/Chengdu/CD80_dataset/Human-Driving and AV Crash Data/Self-Driving Crash Datasets/SGO-2021-01_Incident_Reports_ADAS.csv'

data = pd.read_csv(file_path)
data.head()
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data.info()
```

```python
# Handle missing values
data.fillna('Unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python
# Convert dates to datetime format
data['Incident Date'] = pd.to_datetime(data['Incident Date'], errors='coerce')
data['Notice Received Date'] = pd.to_datetime(data['Notice Received Date'], errors='coerce')
```

```python
# Convert categorical variables to category type
categorical_columns = ['Make', 'Model', 'Crash With', 'City', 'State', 'Weather - Clear', 'Lighting']
for col in categorical_columns:
    data[col] = data[col].astype('category')
```

```python
data.head()
```

```python
# drop rows with unknown values of each column
data = data[data['Make'] != 'Unknown']
data = data[data['Model'] != 'Unknown']
data = data[data['Crash With'] != 'Unknown']
data = data[data['City'] != 'Unknown']
data = data[data['State'] != 'Unknown']
data = data[data['Weather - Clear'] != 'Unknown']
data = data[data['Lighting'] != 'Unknown']

data.info()


```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)

# Encoding categorical variables
data = pd.get_dummies(data, columns=['Make', 'Model', 'Crash With', 'City', 'State'], drop_first=True)

# Normalize numerical columns
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_columns = ['Mileage', 'Posted Speed Limit (MPH)']
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```
